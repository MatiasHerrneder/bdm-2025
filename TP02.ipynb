## Exploracion

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/bdm-unlu/2025/main/TPs/TP02/API_4_DS2_en_csv_v2_3434744.csv')

pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
display(df.head())
```

```python
indicator_counts_with_names = df.groupby(['Indicator Code', 'Indicator Name']).size().reset_index(name='count')
display(indicator_counts_with_names)
```

```python
pd.set_option('display.max_rows', None)
country_code_counts = df['Country Code'].value_counts()
display(country_code_counts)
```

## CSV Paises

```python
df_paises = df[['Country Name', 'Country Code']].drop_duplicates().reset_index(drop=True).rename(columns={
    'Country Name': 'Nombre del Pais',
    'Country Code': 'Codigo del Pais'
})

df_paises['ID'] = df_paises.index + 1
df_paises = df_paises[['ID', 'Nombre del Pais', 'Codigo del Pais']]

df_paises.to_csv('country_data_with_id.csv', index=False)

display(df_paises)
```

## CSV Preguntas

```python
import re

# Get unique indicator names
unique_indicators = df['Indicator Name'].drop_duplicates().reset_index(drop=True)

# Function to remove content within parentheses
def remove_parentheses_content(text):
    return re.sub(r'\s*\(.*?\)\s*', '', text).strip()

# Apply the function to create the 'consulta' column
consulta_series = unique_indicators.apply(remove_parentheses_content)

# Create a new DataFrame
df_consulta = pd.DataFrame({'consulta': consulta_series})

# Add an incremental ID
df_consulta['ID'] = df_consulta.index + 1

# Reorder columns
df_consulta = df_consulta[['ID', 'consulta']]

# Save to CSV
df_consulta.to_csv('consulta_data_with_id.csv', index=False)

display(df_consulta)
print("CSV file 'consulta_data_with_id.csv' created successfully.")
```

## CSV ranking duracion de educacion primaria 2023

```python
# Filter data for the year 2023 and the specific indicator
primary_education_2023 = df[(df['Indicator Code'] == 'SE.PRM.DURS') & (df['2023'].notna())]

# Select relevant columns and sort by duration for ranking
primary_education_ranking = primary_education_2023[['Country Name', '2023']].sort_values(by='2023', ascending=False).reset_index(drop=True)

# Add an incremental ID
primary_education_ranking['Ranking'] = primary_education_ranking.index + 1

# Rename the duration column for clarity
primary_education_ranking = primary_education_ranking.rename(columns={'2023': 'Duration (Years)'})

# Reorder columns
primary_education_ranking = primary_education_ranking[['Ranking', 'Country Name', 'Duration (Years)']]

# Save to CSV
primary_education_ranking.to_csv('primary_education_duration_ranking_2023.csv', index=False)

display(primary_education_ranking)
```
